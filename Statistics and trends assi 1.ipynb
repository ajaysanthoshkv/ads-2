# Apple Quality Analysis

Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import skew
import seaborn as sns
from scipy.stats import kurtosis
df = pd.read_csv('apple_quality.csv')
print(df.head())
```

```python
apple_quality = pd.read_csv('apple_quality.csv',index_col='A_id')
```

```python
apple_quality
```

```python
apple_quality.describe()
```

```python
apple_quality=apple_quality.select_dtypes(exclude=[object])
```

```python
apple_quality.corr()
```

```python
print(skew(apple_quality, axis=0, bias=True))
```

```python
print(kurtosis(apple_quality, axis=0, bias=True))
```

```python
# Create a histogram of a numerical variable
def histo_map():
    plt.hist(apple_quality['Acidity'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Distribution of Data')
    plt.xlabel('Size')
    plt.ylabel('Weight')
    plt.show()
    plt.savefig('histogram.png')
```

```python
histo_map()
```

```python
# Create a line graph of trends over time
def scatter_map():
    plt.scatter(apple_quality['Juiciness'], apple_quality['Ripeness'], color='green')
    plt.title('Juiciness Over Ripeness')
    plt.xlabel('Juiciness')
    plt.ylabel('Ripeness')
    plt.yscale('log')
    plt.show()
    plt.savefig('scatter.png')
```

```python
scatter_map()
```

```python
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
print("Non-numeric columns:", non_numeric_cols)

# Handle non-numeric values
# For example, you can encode categorical variables or handle missing values
# For simplicity, we'll just drop non-numeric columns in this example
df_numeric = df.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = df_numeric.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Create a correlation heatmap
def corr_map():
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()
    plt.savefig('Heatmap.png')
```

```python
corr_map()
```
